```python
import pandas as pd
import numpy as np
import seaborn as sns
import io
import statsmodels.formula.api as smf
import scipy.stats as stats
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/sleep75.csv')
df
```

```python
# Определим датафрейм
sleep_df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/sleep75.csv')
sleep_df['sleep60'] = sleep_df['sleep']/60 # создадим столбец часов времени
# Отфильтруем датафрейм по географическому признаку
sleep_south = sleep_df[sleep_df['south']==1] # Датафрейм для юга
sleep_north = sleep_df[sleep_df['south']==0] # Датафрейм дла севера
sleep_south
```

```python
sns.lmplot(data = sleep_df, x = 'totwrk', y = 'sleep', ci = None, hue = 'south')
```

```python
slp = smf.ols(data= sleep_df, formula='sleep60~totwrk+age+I(age**2)+male +smsa').fit()
slp_sth = smf.ols(data= sleep_south, formula='sleep60~totwrk+age+I(age**2)+male +smsa').fit()
slp_nrth = smf.ols(data= sleep_north, formula='sleep60~totwrk+age+I(age**2)+male +smsa').fit()
```

```python
F_obs =(slp.ssr - (slp_sth.ssr + slp_nrth.ssr) )/((slp_sth.ssr + slp_nrth.ssr)) * (len( sleep_north ) + len( sleep_south ) - 14) / 7
F_obs
```

```python
slp.ssr
```

```python
print(f'RSS:{slp.ssr.round(2)}, RSS_1:{slp_sth.ssr.round(2)},RSS_2:{slp_nrth.ssr.round(2)} ')
```

```python
alpha = 0.05
stats.f.ppf(1-alpha,6,len( sleep_north ) + len( sleep_south ) - 12  )
```

### Вывод:
- $F_{obs} > F_{cr}$
- Данные не согласуются с нулевой гипотезой
- Структурные различия значимы

## Задача 4.3

```python
import pandas as pd
import numpy as np
import seaborn as sns
import io
import statsmodels.formula.api as smf
import scipy.stats as stats
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/wage1.csv')
df
```

```python
# Определим датафрейм
wage_df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/wage1.csv')
wage_df['wage1'] = wage_df['wage'] # создадим столбец часов времени
# Отфильтруем датафрейм по полу
wage_male = wage_df[wage_df['female']==0] # Датафрейм для мужчин
wage_female = wage_df[wage_df['female']==1] # Датафрейм для женщин
wage_male
```

```python
wage_df['log(wage)']  = np.log(wage_df['wage'])
wage_df
```

```python
sns.lmplot(data = wage_df, x = 'exper', y = 'log(wage)', ci = None, hue = 'female')
```

```python
wg = smf.ols(data= wage_df, formula='np.log(wage)~exper+I(exper**2)+smsa +married').fit()
wg_ml = smf.ols(data= wage_male, formula='np.log(wage)~exper+I(exper**2)+smsa +married').fit()
wg_fml = smf.ols(data= wage_female, formula='np.log(wage)~exper+I(exper**2)+smsa +married').fit()
```

```python
print(f'RSS:{wg.ssr.round(2)}, RSS_1:{wg_ml.ssr.round(2)},RSS_2:{wg_fml.ssr.round(2)} ')
```

```python
F_obs =( wg.ssr - (wg_ml.ssr + wg_fml.ssr) )/((wg_ml.ssr + wg_fml.ssr)) * (len( wage_male ) + len( wage_female) - 10) / 5
F_obs
```

```python
alpha = 0.05
stats.f.ppf(1-alpha,5,len( wage_male ) + len( wage_female) - 10 )
```

### Вывод:
- $F_{obs} > F_{cr}$
- Данные не согласуются с нулевой гипотезой
- Структурные различия значимы

```python
df['low'] = (df['Estimate_numb'] - df['Std.error']*t_cr).round(2)
df['upper'] = (df['Estimate_numb'] + df['Std.error']*t_cr).round(2)
df
```
